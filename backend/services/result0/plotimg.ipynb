```python
from PIL import Image
import matplotlib.pyplot as plt


fig, axs = plt.subplots(3, 5, figsize=(20, 10))

for i in range(15):  # This loops from 0 to 14, matching the 3 rows by 5 columns
    img = Image.open(f"route{i+1}.jpg")  # Load each image
    axs[i//5, i%5].imshow(img)          # Display image in the correct subplot
    axs[i//5, i%5].axis("off")          # Hide axes for clarity

fig.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0.01, hspace=0.01)
plt.show()  # Display all the subplots
```
